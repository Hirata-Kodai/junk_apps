```python
import altair as alt
import pandas as pd
import numpy as np
from vega_datasets import data
```

```python
D_transformed_df = pd.DataFrame(np.load("./embed_sample/embed_sweater.npy"), columns=["dim1", "dim2"])
haiku_df = pd.read_csv("./search_result_sample/search_result_sweater.csv")
haikus = haiku_df['本文']
source = pd.concat([D_transformed_df, pd.DataFrame(haikus)], axis=1)
source.columns = ["dim1", "dim2", "haiku"]
source["haiku"] = source["haiku"].astype(str)
print(source.dtypes)
print(source.head())
alt.Chart(source).mark_circle(size=60).encode(
x='dim1',
y='dim2',
tooltip='haiku').interactive()
```
